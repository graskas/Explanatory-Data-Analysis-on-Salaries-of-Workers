## Importing packages

```python
import numpy as np
import pandas as pd
import matplotlib 
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Loading the Dataset

```python
df = pd.read_csv('Salary_Data.csv')
df
```

## Data Cleaning and Preparation

**Understanding the data by checking the shape of the data and data types of the variables**

```python
df.info()
```

```python
##checking for missing values

df.isnull().sum()
```

By checking for the number of missing values, it can be seen that all the variables in the data has a missing value at least.


The missing values would be dropped in this case to facilitate the other parts of this project

```python
df =df.dropna()
```

```python
df.isnull().sum()
```

It can then be seen that there are no more missing values so the rest of the project can be carried on

```python
df.describe()
```

## Data Visualisation 

```python
# Setting plotting style

sns.set_style("whitegrid")
```

**A pie chart to depict the gender proportions**

```python
sex= df["Gender"].value_counts()

labs= ["Male","Female","Others"]
```

```python
plt.pie(sex, labels= labs, autopct= '%1.2f%%');
```

It can be seen that most of the respondents are males with less than 0.5% of them in the others category

**Age Distribution of Respondents**

```python
plt.figure(figsize=(12,6))
plt.title("Age Distribution of Respondents")
plt.xlabel("Age")
plt.ylabel("Number of respondents")

plt.hist(df.Age);
```

The histogram above depicts that the Age data is skewed to the right which means that most of the respondents were between the ages of 20 and 40 years and few above the 40 year mark.

**A bar plot showing the Top 10 job titles** 

```python
top_job_titles= df["Job Title"].value_counts().head(10)
top_job_titles
```

```python
plt.figure(figsize= (17,8))
sns.barplot(x= top_job_titles, y=top_job_titles.index )


plt.title("Job Title")
plt.xlabel("Number of Respondents")
plt.ylabel(None);
```

**A scatterplot showing the relationship between the Age of the worker and the amount of Salary they earn**

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x = "Age", y= "Salary", data = df)

plt.title("Scatterplot of Age and Salary");
```

It can be seen that there is a positive relationship between the Age of the respondent and the salary of the respondent. That is as the age of the individual increases, the salary of the individual also increases.

**A scatter plot showing the relationship between years of experience and Age**

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x = "Years of Experience", y= "Salary",data = df)

plt.title("Scatterplot of Years of Experience and Salary");
```

It can be seen that there is a positive relationship between the Years of experience of the respondent and the salary of the respondent. That is as the years of experience of the individual increases, the salary of the individual also increases.
